```python
from ipywidgets import FileUpload
from IPython.display import display
```

```python
uploader = FileUpload(accept='.jpg', multiple=False)
display(uploader)
```

```python
from pprint import pprint
print(uploader.value)
```

```python
import base64

content = uploader.value['pexels-k-prodan-421922755-29243214.jpg']['content']

img_bytes = bytes(content)
img_b64 = base64.b64encode(img_bytes).decode('utf-8')
```

```python
from google.colab import userdata
from langchain.messages import HumanMessage, SystemMessage
from langchain.agents import create_agent
from langchain_google_genai import ChatGoogleGenerativeAI

multimodal_questions = [
    SystemMessage(content="You are intelligent assistant who help in analysing the given image and give a brief summary of it"),
    HumanMessage(content=[
    {"type": "text", "text": "Tell me about this picture"},
    {"type": "image", "base64": img_b64, "mime_type": "image/jpg"}])
    ]

model = ChatGoogleGenerativeAI(model="gemini-2.5-flash", api_key=userdata.get("GOOGLE_API_KEY"))
model
```

```python
response = model.invoke(multimodal_questions)
print(response.content)
```
